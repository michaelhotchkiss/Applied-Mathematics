## Problem 1

```python
import numpy as np
from sympy import diff, cos, sin, symbols,exp,Function
x,y,t = symbols('x y t')
x_1 = cos(x**3)
x_1_d = diff(x_1,x)
print('1')
print(f'a) {x_1_d}')

x_2 = x**x
x_2_d = diff(x_2,x)
print(f'b) {x_2_d}')

x_3 = exp(sin(2*x)) * cos(x)
x_3_d = diff(x_3,x)
print(f'c) {x_3_d}')

x_t = Function('x')(t)
y_t = Function('y')(t)

eq_4 = cos(x_t**2 + y_t**2)
eq_4_d = diff(eq_4,t)
print(f'd) {eq_4_d}')
```

## Problem 2

```python
from sympy import exp,symbols,solve,nsolve
import numpy as np
import matplotlib.pyplot as plt
from math import log
lamb,t  = symbols('lamb t')
c = 2

x_t = c  * exp(lamb * t)
print('a')
print(x_t)
lamb_array = np.array([-5,-1,0,.01,.1])
times = np.linspace(0,5)

keys = np.array(['lambda 1', 'lambda 2','lambda 3','lambda 4','lambda 5'])

for lambs in lamb_array:
    x_t = c  * np.exp(lambs * times)
    plt.plot(times,x_t)

plt.legend(keys)
plt.xlabel('t')
plt.ylabel('x_t')
print('b')
plt.show()

t = symbols('t',real=True)
time = (4.468*10**9)
lamb = log(2)/time
print('c')
print(f'lambda = {lamb}')
x_t = 100 * exp(-lamb * t)
soln = nsolve(100 * exp(-lamb * t)- 5,t,2)
print('d')
print(f'{soln} years')
```

## Problem 3

```python
from sympy import symbols, diff, sin, lambdify, pprint

x = symbols('x')
eq = sin(x)/x
x_d = diff(eq,x)
x_dd = diff(x_d,x)
x_ddd = diff(x_dd,x)

eq_l = lambdify(x,eq)
x_d_l = lambdify(x,x_d)
x_dd_l = lambdify(x,x_dd)
x_ddd_l = lambdify(x,x_ddd)
expansion = eq_l(0) + x_d_l(0) * x + x_dd_l(0) * x + x_ddd_l(0) * x
pprint(expansion)
```
