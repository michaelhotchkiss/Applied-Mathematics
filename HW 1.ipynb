## Problem 1

```python
import numpy as np
from sympy import diff, cos, sin, symbols,exp,Function
x,y,t = symbols('x y t')
x_1 = cos(x**3)
x_1_d = diff(x_1,x)
print('1')
print(f'a) {x_1_d}')

x_2 = x**x
x_2_d = diff(x_2,x)
print(f'b) {x_2_d}')

x_3 = exp(sin(2*x)) * cos(x)
x_3_d = diff(x_3,x)
print(f'c) {x_3_d}')

x_t = Function('x')(t)
y_t = Function('y')(t)

eq_4 = cos(x_t**2 + y_t**2)
eq_4_d = diff(eq_4,t)
print(f'd) {eq_4_d}')
```

## Problem 3

```python
from sympy import symbols, diff, sin, lambdify
import numpy as np
import matplotlib.pyplot as plt
x = symbols('x')

eq = sin(x)
x_d = diff(eq,x)
x_dd = diff(x_d,x)
x_ddd = diff(x_dd,x)
x_4d = diff(x_ddd,x)
x_5d = diff(x_4d,x)

eq_l = lambdify(x,eq)
x_d_l = lambdify(x,x_d)
x_dd_l = lambdify(x,x_dd)
x_ddd_l = lambdify(x,x_ddd)
x_4d_l = lambdify(x,x_4d)
x_5d_l = lambdify(x,x_5d)

expansion = eq_l(0)/x + x_d_l(0)/x * x + x_dd_l(0)/x * x**2/np.math.factorial(2) + x_ddd_l(0)/x * x**3/np.math.factorial(3) + x_4d_l(0)/x * x**4/np.math.factorial(4) + x_5d_l(0)/x * x**5/np.math.factorial(5)
expansion_l = lambdify(x,expansion)

print('a')
print(f'Taylor expansion = {expansion}')

times = np.linspace(-5,5)
expansion_values = expansion_l(times)
real_values = np.sin(times)/times

plt.plot(times, expansion_values)
plt.plot(times,real_values)
plt.legend(['taylor', 'real'])
plt.show()
```

```python
from sympy import symbols, diff, sin, lambdify
import numpy as np
import matplotlib.pyplot as plt
x = symbols('x')

eq = 3**x
x_d = diff(eq,x)
x_dd = diff(x_d,x)
x_ddd = diff(x_dd,x)

eq_l = lambdify(x,eq)
x_d_l = lambdify(x,x_d)
x_dd_l = lambdify(x,x_dd)
x_ddd_l = lambdify(x,x_ddd)

expansion = eq_l(0) + x_d_l(0) * x + x_dd_l(0) * x**2/np.math.factorial(2)
print('b')
print(f'Taylor expansion = {expansion}')
expansion_l = lambdify(x,expansion)
times = np.linspace(-5,5)
expansion_values = expansion_l(times)
real_values = 3**times

plt.plot(times, expansion_values)
plt.plot(times,real_values)
plt.legend(['taylor', 'real'])
plt.show()
```
